```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.describe()
```

```python
iris.isnull().sum()
```

```python
iris.dtypes
```

```python
iris.corr()
# checking the relationship  between numeric features
```

```python
sns.heatmap(iris.corr(), annot=True)
```

```python
sns.jointplot(iris.sepal_length,iris.petal_length, kind='hex')
# for one bivariate analysis 
```

```python
sns.jointplot(iris.sepal_width,iris.petal_width, kind='reg')
```

```python
sns.pairplot(data=iris)
# this plot is use for the relationship between more than one numeric feature
```

```python
iris.head(2)
```

```python
sns.pairplot(data=iris, hue='species')
```

```python
sns.pairplot(data=iris,)
```

```python
iris['species'].value_counts()
```

```python
sns.displot(data=iris, x='sepal_length', kde=True)
# this is use for one feature distribution 
```

```python
sns.displot(data=iris, x='sepal_length', kde=False, hue='species')
```

```python
sns.displot(data=iris, x='petal_length', kde=False, hue='species')
```
